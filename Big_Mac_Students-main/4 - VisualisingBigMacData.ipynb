```python
import matplotlib.pyplot as plt
import pandas as pd

#Read in the big_mac_update dataset to big_mac_df DataFrame
bigmacdata = pd.read_csv('data/big_max_aud_update.csv')

#Create the plot using big_mac_df
bigmacdata.plot(
                kind='scatter',
                x='AUD Price',
                y ='Country',
                color='orange',
                alpha=0.3,
                title='Big Mac Price across different Countries'
                )

#wanted this to be a bar but couldnt make it work.
#Show the plot
plt.show()
```
